#### Author & Narrator column
###### To make these easy to read and understand, we are getting these features rid of their unnecesary prefixes and proving them spaces between the first and the last name.

```python
# Remove the 'Writtenby:' prefix from the 'author' column.
data['author'] = data['author'].str.replace('Writtenby:', '', regex=False)
data['author'].head()

# Function to add spaces between names
def add_spaces(names):
    spaced_names = []
    for name in names:
        spaced_name = ''
        for i, char in enumerate(name):
            # Add space before uppercase letters
            if i > 0 and char.isupper() and name[i - 1] not in ["'", '"']:
                spaced_name += ' '
            spaced_name += char
        spaced_names.append(spaced_name)
    return spaced_names

data['author'] = add_spaces(data['author'])
```

```python
#Removing the 'narrated by:' suffix from 'narrator' column...
data['narrator'] = data['narrator'].str.replace('Narratedby:', '', regex=False)
data['narrator'].head()
# Function to add spaces between narrator names
def add_spaces(names):
    spaced_names = []
    for name in names:
        spaced_name = ''
        for i, char in enumerate(name):
            # Add space before uppercase letters (excluding the first one and after a single quote)
            if i > 0 and char.isupper() and name[i - 1] not in ["'", '"']:
                spaced_name += ' '
            spaced_name += char
        spaced_names.append(spaced_name)
    return spaced_names

data['narrator'] = add_spaces(data['narrator'])
```

#### Time Column
###### The time feature in the dataset represents the duration of the audiobook in string format, here we will standardize  it to numerical value in minutes for easier analysis.

```python
def convert_time_to_minutes(time_str):
    # Use regex to find numbers followed by 'hrs' or 'mins'
    hours = 0
    minutes = 0
    hours_match = re.search(r'(\d+)\s*hrs', time_str)
    minutes_match = re.search(r'(\d+)\s*mins', time_str)

    if hours_match:
        hours = int(hours_match.group(1))
    if minutes_match:
        minutes = int(minutes_match.group(1))

    return hours * 60 + minutes


data['time'] = data['time'].apply(convert_time_to_minutes)

data['time'].head()
```

#### Release Date 
###### Similar to the above, we also change the datatype of the releasedate feature to datetime format.

```python
#Release Date: This is currently a string; converting it to a datetime format would be more appropriate for any time-based analysis.
data['releasedate'] = pd.to_datetime(data['releasedate'], format='%d-%m-%y', errors='coerce')
data['releasedate']
```

#### Stars
###### This column has two features: 1. The Ratings  & 2. Number of Ratings
###### So, we have to split the column 
###### Now getting on to the stars feature, in order to standardize it and make it beneficial for analysis we convert it to numerical values.

```python
data['stars'].str.replace('Not rated yet','0')
data['Total Ratings']=data['stars'].str.extract(r'(\d+)( *ratings*)')[0]
data['Total Ratings']=data['Total Ratings'].astype('Int64')
data['Total Ratings']=data['Total Ratings'].fillna(0)

data['stars']=data['stars'].str.extract(r'([0-9]\.*\d*)').astype('Float64')
data['stars']=data['stars'].fillna(0)
```

```python
data.head()
```

#### Price
###### Furthermore we also standardize the price feature by replacing and 'free' item price with numerical zero and converting datatype to float.

```python
# Replace 'Free' with '0' and remove any commas
data['price'] = data['price'].replace('Free', '0').str.replace(',', '')

# Convert the 'price' column to float
data['price'] = pd.to_numeric(data['price'], errors='coerce')

# You can then check the output with
data['price']
```

###### Moreover, to make the dataset look better we capitalize the first letter of the features and values of the language features, also renaming some of the column names.

```python
# Capitalize the first letter of each column name
data.columns = [column.capitalize() for column in data.columns]
data.rename(columns={
    'Stars': 'Stars(out of 5)',
    'Time': 'Time(in minutes)'
}, inplace=True)
data.head()
```

#### Box Plot for Total Ratings vs Price
We plot Price vs total rating, this will help us understand if more expensive audiobooks tend to recieve more ratings or not. We can see from this that mid range audiobooks have higer ratings.

```python
# Create price bins
data['Price Bins'] = pd.cut(data['Price'], bins=[0, 200, 400, 600, 800, 1000, 1200, 1400, 1600, 1800, 2000, max(data['Price'])])

# Create a box plot for Total Ratings vs. Price Bins
plt.figure(figsize=(14, 8), dpi=80)
sns.boxplot(data=data, x='Price Bins', y='Total ratings')
plt.title('Distribution of Total Ratings Across Different Price Ranges')
plt.xlabel('Price Bins')
plt.ylabel('Total Ratings')
plt.xticks(rotation=45)
plt.show()
```

#### Scatter Plot for Price vs Stars

### Three-Variable Visualisation

#### Scatter Plot between audiobook duration, star rating and total number of ratings
We plot audiobook duration, star rating and total number of rating as intesity of the colour, we can see from the figure that while lengthier audiobooks tend to be centred around average ratings, shorter ones show a wide range of ratings. The colour intensity of the overall scores, however, does not clearly correspond with audiobook quality, suggesting a sophisticated grading methodology.

```python
# Sample 10,000 entries from the dataset
sampled_data = data.sample(n=10000, random_state=1)
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(sampled_data['Time(in minutes)'], sampled_data['Stars(out of 5)'],
                      c=sampled_data['Total ratings'], cmap='viridis', alpha=0.6)
plt.colorbar(scatter, label='Total Ratings')
plt.xlabel('Time in minutes')
plt.ylabel('Stars out of 5')
plt.title('Scatter Plot of Time vs. Stars Colored by Total Ratings')
plt.grid(True)
plt.show()
```

#### Buble Chart between audiobook duration, star rating and Prices.
We plot audiobook duration, star rating and prices depending on the buble size. Although we can see from the figure that there is no clear correlation. There are expensive audiobooks having high ratings as well as low and vice versa. Therefore, price doesn't seem to be a strong indicator of either duration or rating.

```python
plt.figure(figsize=(10, 6))
bubble = plt.scatter(sampled_data['Time(in minutes)'], sampled_data['Stars(out of 5)'],
                     s=sampled_data['Price'] / 10, alpha=0.6, edgecolors='w', linewidth=0.5)
plt.xlabel('Time in minutes')
plt.ylabel('Stars out of 5')
plt.title('Bubble Chart of Time vs. Stars with Bubble Size as Price')
plt.grid(True)
plt.show()
```

#### 3D Scatter Plot between audiobook duration, prices and total ratings.
We plot audiobook duration, prices and total ratings. Here there is lack of a clear relationship between price, length, and total ratings suggests that ratings aren't a reliable indicator of any of those two variables. However, longer audiobooks typically receive higher amount of ratings.

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter3d = ax.scatter(sampled_data['Time(in minutes)'], sampled_data['Price'], sampled_data['Total ratings'],
                       c=sampled_data['Total ratings'], cmap='coolwarm', alpha=0.6)
fig.colorbar(scatter3d, ax=ax, label='Total Ratings')
ax.set_xlabel('Time in minutes')
ax.set_ylabel('Price')
ax.set_zlabel('Total Ratings')
ax.set_title('3D Scatter Plot of Time, Price, and Total Ratings')
plt.show()
```
